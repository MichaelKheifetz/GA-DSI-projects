```python
%matplotlib inline
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

theta = 3
x = np.random.poisson(theta, size=1000)
x.mean(), x.var()

x= plt.hist(x, bins = 30)
```

```python
y= stats.poisson.pmf(range(10), theta)
z = plt.plot(y, 'ro')
```

```python
from scipy.misc import comb
pbinom = lambda y, n, p: comb(n,y) * p ** y * (1-p)**(n-y)
pbinom (2, 3, 5)
```

```python
pvals = np.linspace(0,1)
y = 5
plt.plot(pvals, [pbinom(y, 10, p) for p in pvals])
```

```python
# simulate gamma means
r = 3
p = 0.7

lan = np.random.gamma(r, p/(1-p), size = 100)

#simulate sample Poisson conditional on lambda
simulation_values = np.random.poisson(lan)
simulation_values.sum()/100.
```
