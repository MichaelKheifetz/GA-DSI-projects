## Measure Theory

We already spoke about sets and subsets yesterday. Two important examples that we spoke about were the set of all real numbers denoted by $\mathbb{R}$ and the set of all integers $\mathbb{N}$. 

*Let's look at the following sets:*

**Example 1**
$$ A = \{ n \in \mathbb{N} \,\, |\,\,  n\%2 == 0 \} \subset \mathbb{N}$$
The set $A$ is the subset of $\mathbb{N}$ that consists of all even numbers.

**Example 2**
$$ B = \{ x \in \mathbb{R} \,\, | \,\, -1 \leqslant x < 1\} \subset \mathbb{R}$$
The set $B$ is the subset of $\mathbb{R}$ that contains all numbers between -1 and 1 excluding 1.

*Let's talk about all the things that we can do with two sets.*
1. $A \cup B$ is the **union** of $A$ and $B$ and consists of all elements that are either in $A$ or in $B$ (OR Logic)
2. $A \cap B$ is the **intersection** of $A$ and $B$ and contains all elements that are in $A$ and $B$ (AND Logic)
3. $A \setminus B$ is the **difference** and consists of all points in $A$ that are not in $B$.
4. Two sets $A$ and $B$ are called disjoint if $A \cap B = \emptyset $ which means that any point in A is not in B and vice versa.
5. Let $A \cap B = \emptyset$ be two disjoint sets. We call $A \uplus B$ the **disjoint union** of A and B. (This notation means the same as 1. We are only emphasising that A and B are disjoint)

Let's take any set that we would like and call it $\mathbb{A}$. We can now look at a very special set namely the set of all subsets of $\mathbb{A}$ which is denoted by 
$$ \mathbb{P} (\mathbb{A}) = \{ \text{All subsets of } \mathbb{A} \}$$

**Example 1**

Consider $\mathbb{R}$ the set of all real numbers. The two sets defined by $A_1 = \{ x \, | \, -1 \leqslant x < 1 \}$ and $A_2 = \{ x \, | \, -2 \leqslant x < 1 \}$ are both subsets of $\mathbb{R}$ and thus $A_1,A_2 \in \mathbb{P}(\mathbb{R})$.

## Definition of Measures

A measure of a set $\mathbb{A}$ is a mapping that maps every subset of $A$ onto a positive real number
$$\mu : \mathbb{P}(\mathbb{A}) \to \mathbb{R}^+$$
such that
1. $\mu(S) \geqslant 0$ for every subset $S \in \mathbb{P}(\mathbb{A})$
2. $\mu(\emptyset) = 0$
3. $\mu(S_1 \cup S_2) \leqslant \mu(S_1) + \mu(S_2)$ for every subset $S _1, S_2 \in \mathbb{P}(\mathbb{A})$
4. $\mu(S_1 \uplus S_2) = \mu(S_1) + \mu(S_2)$ for two disjoint subsets $S _1, S_2 \in \mathbb{P}(\mathbb{A})$

$\mu : \mathbb{P}(\mathbb{A}) \to \mathbb{R}^+$ is called a **Propability Measure** if
$$ \mu(\mathbb{A}) = 1$$

We will generally denote a Propability Measure with a P.

**Example 1**

Let's look at the set of all integers $\mathbb{N}$ and define the following measure:
$$\mu: \mathbb{P}(\mathbb{N}) \to \mathbb{R}^+ : \mu(S) = \text{Number of elements in } S$$ 

1. $S = \{0,1,2,3,4,5,6\}$ then $\mu(S) = \mu(\{0,1,2,3,4,5,6\}) = 7$
2. $S_1 = \{0,1,2,3,4,5,6\}$ and $S_2 = \{\text{even numbers}\}$ then $\mu(S_1 \cap S_2) = \mu(\{2,4,6\}) = 3$

**Example 2**

Let's look at the set of all real numbers $\mathbb{R}$ and define the following measure:
$$\lambda: \mathbb{P}(\mathbb{R}) \to \mathbb{R}^+ : \lambda(S) = \text{Length of the interval } S$$ 

1. $S = \{x \in \mathbb{R} \, | \, 2 \leqslant x < 4 \}$ then $\mu(S) = \lambda([2,4)) = 4-2 = 2$
2. $S_1 = \{x \in \mathbb{R} \, | \, 2 \leqslant x < 4 \}$ and $S_2 = \{x \in \mathbb{R} \, | \, 6 \leqslant x < 8 \}$ then $\lambda(S_1 \uplus S_2) = \lambda(S_1) + \lambda(S_2) = 2 + 2 = 4$
3. The set of all integers $\mathbb{N}$ is a subset of the set of all real numbers $\mathbb{R}$. Thus we can calculate $\lambda(\mathbb{N})$
$$ \lambda(\mathbb{N}) = \lambda(\uplus_{k = 0}^{\infty} \{k\}) = \sum_{k = 0}^{\infty} \lambda(\{k\}) =0 + 0 + 0 + ... = 0 $$

## Indicator Functions

Let $S$ be any subset of another another set $\mathbb{A}$. We could for example choose the set of real numbers $\mathbb{R}$ and the intervall $S = [0,1)$.
We define the indicator function over $S$ 
$$ \mathbb{1}_{S}(x) = 1 \text{ if } x \in S \text{ else } 0 $$ 

**Examples**
1. $\mathbb{1}_{[0,1)}(0.5) = 1$
2. $\mathbb{1}_{[0,1)}(2) = 0$
3. $\mathbb{1}_{[0,1) \cup [2,4)}(3) = 1$
4. $\mathbb{1}_{\{even\}} (3) = 0$ and $\mathbb{1}_{\{even\}} (2) = 1$

````python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
````

````python
def I(l,u):
    return lambda x: 1 if x < u and x >= l else 0

def createFunc(l,u,s,F):
    labels = np.arange(l,u+s,s)
    return lambda x: sum([np.min([F(labels[i]), F(labels[i+1])]) * I(labels[i],labels[i+1])(x) for i in range(len(labels) -1)])
````

````python
F = lambda x: (1)**2 * I(-2,-1)(x) + (0)*I(-1,0)(x) +(0)*I(0,1)(x) +\
              (1)**2 * I(1,2)(x) + 2**2*(I(-3,-2)(x) + I(2,3)(x))
F1 = lambda x: np.exp(-x**2)
X = np.arange(-3,3,0.01)
FX = lambda s: [createFunc(-3,3,s,F1)(x) for x in X]
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(2,2,1)
ax1.plot(X, FX(1), label = 'step size:1')
ax1.plot(X, F1(X), label = 'F(x) = Exp(-x^2)')
ax1.legend(loc = 'best', fontsize = 'small')
ax2 = fig.add_subplot(2,2,2)
ax2.plot(X, FX(0.75), label = 'step size:0.75')
ax2.plot(X, F1(X), label = 'F(x) = Exp(-x^2)')
ax2.legend(loc = 'best', fontsize ='small')
ax3 = fig.add_subplot(2,2,3)
ax3.plot(X, FX(0.5), label = 'step size:0.5')
ax3.plot(X, F1(X), label = 'F(x) = Exp(-x^2)')
ax3.legend(loc = 'best', fontsize = 'small')
ax4 = fig.add_subplot(2,2,4)
ax4.plot(X, FX(0.25), label = 'step size:0.25')
ax4.plot(X, F1(X), label = 'F(x) = Exp(-x^2)')
ax4.legend(loc = 'best', fontsize = 'small')
````

## Integration

We can now define the Integral of an Indicator function over a subset A with repsect to any measure $\mu$ in the following way

$$ \int_{A} d\mu(x) \,\, (b_1*\mathbb{1}_{B_1}(x) + b_2*\mathbb{1}_{B_2}(x) ) = b_1* \mu(A \cap B_1) + b_2* \mu(A \cap B_2)$$

** Examples **
1. $$ f(x) = 0.5*\mathbb{1}_{[-1,1)}(x) + 2*\mathbb{1}_{[2,3)}(x)$$ 
$$ \int_{[-2,2.5)} d\lambda(x) \,\, f(x) = 0.5 * \lambda([-1,1)) + 2 *\lambda([2,2.5)) = 0.5*2 + 2* 0.5 = 1$$
2. $$F(n) = \sum_{k = 0}^{\infty} 0.5^k \mathbb{1}_{\{k\}}(n) \text{ and thus } \int_{\mathbb{N}} d\mu(n) \,\, F(n) = \sum_{k = 0}^{\infty} 0.5^k = \frac{1}{1-0.5} = 2$$

We can integrate any function F by approximating it with indicator functions. We can thus now write something like this:

$$ \int_{[-1,1)} d\lambda(x) \, Exp(-x^2)$$

In the typical setting where we are looking at functions from $F: \mathbb{R} \to \mathbb{R}$ the integral is the area between the graph and the x-axis.

````python
X = np.arange(-5,5.1,0.1)
X2 = np.arange(-1,1.1,0.1)
F = lambda x: np.exp(-x**2)
fig = plt.figure(figsize=(6,3))
ax = fig.add_subplot(1,1,1)
ax.plot(X,F(X), color = 'blue', label = None)
ax.fill_between(X2, F(X2), color = 'blue', alpha = 0.3, label = 'Area between -1 and 1')
ax.legend(loc = 'best', fontsize = 'small')
ax.set_xlabel('x')
ax.set_ylabel('F(x)')
````

The Scipy Package includes functions which allow us to integrate. 
``` python
(area, precision) = scipy.integrate.quad(func, lowerBound, upperBound)
```

````python
from scipy import integrate
integrate.quad(F, -np.inf.np.inf)
````

We can also use the integral function to create a new function for example

$$ F(x) = \int_{-\infty}^{x} d\lambda(t) \, Exp(-t^2)$$

````python
intF = lambda x: integrate.quad(F, -np.inf, x)[0]
intFX = [intF(x) for x in X]
fig = plt.figure(figsize=(6,3))
ax = fig.add_subplot(1,1,1)
ax.plot(X,F(X), color = 'blue', label = None)
ax.plot(X, intFX, color = 'red', label = 'Int of F')
ax.fill_between(X2, F(X2), color = 'blue', alpha = 0.3, label = 'Area between -1 and 1')
ax.legend(loc = 'best', fontsize = 'small')
ax.set_xlabel('x')
ax.set_ylabel('F(x)')
````
