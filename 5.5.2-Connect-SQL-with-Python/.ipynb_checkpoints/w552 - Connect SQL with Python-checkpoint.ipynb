We can now combine the power of SQL and Python to read and write to and from a SQL Server.

```python
import string

#Function that takes a connection, a tablename and a list of columns (col_name, col_type) and returns 1 if the table was created and 0 else.

def createTable(conn, tablename, col):
    sqlstr = 'CREATE TABLE ' + tablename + ' ('
    columns = string.join(['{0} {1},'.format(c[0], c[1]) for c in col])
    cursor = conn.cursor()
    sqlstr = sqlstr + columns[:-1] + ')'
    try:
        cursor.execute(sqlstr)
        conn.commit()
        return 1
    except:
        conn.rollback()
        return 0
```
