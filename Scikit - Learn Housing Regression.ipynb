```python
price = [x for x in sac.columns if x == 'price' or x == 'sq_ft' or x == "zip"]


sac.ix[:, price] = (sac.ix[:, price] - sac.ix[:, price].min())\
                    / (sac.ix[:, price].max() - sac.ix[:, price].min())
    
sac
```

```python
y = pd.get_dummies(sac["beds"])
```

```python
# Split the data into training/testing sets
homes_X_train = sac[:-20]
homes_X_test = sac[-20:]

# Split the targets into training/testing sets
homes_y_train = sac.target[:-20]
homes_y_test = sac.target[-20:]

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(homes_X_train, homes_y_train)

# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print("Mean squared error: %.2f"
      % np.mean((regr.predict(homes_y_test) - homes_y_test) ** 2))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(homes_y_test, homes_y_test))

# Plot outputs
plt.scatter(homes_X_test, homes_y_test,  color='black')
plt.plot(homes_X_test, regr.predict(homes_X_test), color='blue',
         linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```
